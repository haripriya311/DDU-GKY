## BOXPLOT

```python
for i in data.columns:
    if data[i].dtype !='object':
        sns.boxenplot(y=data[i],color='red')
        plt.show();
```

```python
def outliers_limit(col):
    Q3,Q1=np.nanpercentile(col,[75,25])
    IQR=Q3-Q1
    UL=Q3+1.5*IQR
    LL=Q1-1.5*Q1
    return UL,LL
```

```python
for column in data.columns:
    if data[column].dtype != "object":
        UL,LL=outliers_limit(data[column])
        data[column]=np.where((data[column]>UL) | (data[column]<LL),np.nan,data[column])
```

```python
for i in data.columns:
    if data[i].dtype !="object":
        sns.boxenplot(y=data[i],color='green')
        plt.show();
```
